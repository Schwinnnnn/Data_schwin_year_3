```python
df1 = pd.read_csv("https://github.com/Schwinnnnn/Data_schwin_year_3/raw/main/Bi/project2/average_revenue_per_household.csv")
df1
```

```python
df1['SOCIO_ECONOMIC'].replace({'รวมสถานะทางเศรษฐสังคมของครัวเรือน': 'Household status aggregation'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ถือครองทำการเกษตร การปลูกพืช การเลี้ยงสัตว์ และการเพาะเลี้ยงสัตวน้ำ ซึ่งที่ดินส่วนใหญ่เป็นของตนเอง': 'The landowner engages in agriculture on their own land'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ถือครองทำการเกษตร การปลูกพืช การเลี้ยงสัตว์ และการเพาะเลี้ยงสัตวน้ำ ซึ่งที่ดินส่วนใหญ่เช่าผู้อื่น/ที่สาธารณะ/อื่น ๆ': 'The landowner engages in agriculture on land that leased from others'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ทำการประมง ป่าไม้ ล่าสัตว์ หาของป่า และบริการทางการเกษตร': 'The individual engages in fishing, forestry and agricultural services'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ดำเนินธุรกิจของตนเองที่ไม่ใช่การเกษตร': 'The self-employed individual unrelated to agriculture'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ปฏิบัติงานวิชาชีพ นักวิชาการ และนักบริหารที่รับจ้าง': 'The professional worker is employed on a contractual basis'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'คนงานเกษตร': 'The agricultural laborer'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'คนงานทั่วไป': 'General laborer'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'เสมียนพนักงาน พนักงานขายและให้บริการ': 'Sales and service staff'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ปฏิบัติงานในกระบวนการผลิต': 'Production workers'}, inplace=True)
df1['SOCIO_ECONOMIC'].replace({'ผู้ไม่ได้ปฏิบัติงานในเชิงเศรษฐกิจ': 'Non-economic workforce'}, inplace=True)
df1
```

```python
color = ["lightgray", "lightgray", "red", "red", "red", "red"]
# Assuming df1 is your DataFrame
# Assuming color is predefined

# Create a function to update the plot based on the selected socio-economic group and column
def update_plot(selected_socio_economic, selected_column):
    plt.clf()  # Clear the current figure

    df_filtered = df1[df1['SOCIO_ECONOMIC'] == selected_socio_economic]

    # Group by 'SOCIO_ECONOMIC' and 'year', then calculate the mean income for the selected column
    mean_income_by_socioeconomic = df_filtered.groupby(['SOCIO_ECONOMIC', 'YEAR'])[selected_column].mean().reset_index()

    # Plotting the bar chart with specified colors
    for socio_economic_group, group_data in mean_income_by_socioeconomic.groupby('SOCIO_ECONOMIC'):
        plt.bar(group_data['YEAR'], group_data[selected_column], label=socio_economic_group, color=color)

    # Adding line graph
    plt.plot(mean_income_by_socioeconomic['YEAR'], mean_income_by_socioeconomic[selected_column], color='blue', marker='o', label='Line Graph')

    plt.xlabel('Year')
    plt.ylabel(f'Mean {selected_column}')
    plt.title(f'Mean {selected_column} by Socio-Economic Group Over the Years')
    plt.show()

# Get unique values of 'SOCIO_ECONOMIC' column
socio_economic_values = df1['SOCIO_ECONOMIC'].unique().tolist()

# Get unique values of the columns you want to include in the dropdown
column_values = ['MONTHLY_INCOME', 'CURRENT_INCOME', 'MONEY_INCOME', 'FROM_WORK', 'CURRENT_TRANSFER', 'PROPERTY_INCOME', 'NONMONEY_INCOME', 'NONCUR_M_INCOME']

# Create dropdown widgets for socio-economic group and column selection
dropdown_socio_economic = widgets.Dropdown(options=socio_economic_values, description='Socio-Economic:')
dropdown_column = widgets.Dropdown(options=column_values, description='Select Column:')

# Display the dropdown widgets
display(dropdown_socio_economic)
display(dropdown_column)

# Create event handlers to update the plot when the dropdown values change
def on_dropdown_change(change):
    update_plot(dropdown_socio_economic.value, dropdown_column.value)

# Attach the event handlers to the dropdown widgets
dropdown_socio_economic.observe(on_dropdown_change, names='value')
dropdown_column.observe(on_dropdown_change, names='value')

# Initialize the plot with the first values in the dropdown lists
update_plot(socio_economic_values[0], column_values[0])
```

<p>ข้อมูลหลักที่ใช้ในการวิเคราะห์คือรายได้เฉลี่ยต่อเดือนของครัวเรือน<break></break>โดยแยกตามแหล่งที่มาของรายได้ ในช่วงปี พ.ศ. 2554-2564 เราให้ความสำคัญกับรายได้ที่ได้มาจากการทำงาน เนื่องจากมีความสำคัญต่อความเจริญเติบโตของครัวเรือนมากที่สุด จากการวิเคราะห์กราฟพบว่ารายได้เฉลี่ยที่ได้มาจากการทำงานต่อครัวเรือนตั้งแต่ปี พ.ศ. 2558 มีแนวโน้มลดลงอย่างต่อเนื่อง ซึ่งเป็นสัญญาณที่บ่งบอกถึงความเสี่ยงและความทนทานของเศรษฐกิจในระยะยาว</p>

</div>

```python
df3=pd.read_csv('https://github.com/Schwinnnnn/Data_schwin_year_3/raw/main/Bi/project2/euromonitor.csv',skiprows=5, index_col='Category')
df3 = df3.drop(df3.index[44:])
df3.drop(columns=['Geography','Data Type','Unit','Current Constant','2022','2023'], inplace=True)
df3
```

```python
filtered_df = df3.loc[df3.index.str.contains('Real GDP Growth|Inflation')]
# Extract the category from the index using a regular expression
filtered_df['Category'] = filtered_df.index.str.extract('(Real GDP Growth|Inflation)', expand=False)
# Reset the index without adding it as a new column
filtered_df.reset_index(drop=True, inplace=True)

# Display the DataFrame without the index column
melted_data = filtered_df.melt(id_vars='Category', var_name='Year', value_name='Percent Growth')
melted_data
```

```python
import matplotlib.pyplot as plt

# Convert 'Year' column to numeric
melted_data['Year'] = melted_data['Year'].astype(int)

# Convert 'Percent Growth' column to numeric
melted_data['Percent Growth'] = melted_data['Percent Growth'].astype(float)

# Plot
plt.figure(figsize=(10, 6))
ax = plt.gca()

# Plot 'Real GDP Growth'
melted_data[melted_data['Category'] == 'Real GDP Growth'].plot(kind='line', x='Year', y='Percent Growth', 
                                                               marker='o', label='Real GDP Growth', ax=ax)

# Plot 'Inflation'
melted_data[melted_data['Category'] == 'Inflation'].plot(kind='line', x='Year', y='Percent Growth', 
                                                         marker='o', label='Inflation', ax=ax)

# Customize plot
plt.title('Trend of Real GDP Growth and Inflation (2011-2021)')
plt.xlabel('Year')
plt.ylabel('Percent Growth')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='black', linestyle='-', linewidth=.5)

# Show plot
plt.tight_layout()
plt.show()
```
