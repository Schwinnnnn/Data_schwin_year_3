```python
df1 = pd.read_csv("https://github.com/Schwinnnnn/Data_schwin_year_3/raw/main/Bi/project2/cost_per_household.csv")
df1
```

```python
df1.info()
```

```python
df1['TYPE_EXPENSE'].replace({'ค่าใช้จ่ายทั้งสิ้นต่อเดือน': 'Total expenses per month'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'ค่าใช้จ่ายเพื่อการอุปโภคบริโภค': 'Consumption expenses'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'อาหาร และเครื่องดื่ม (ไม่มีแอลกอฮอล์)': 'Food and drink (no alcohol)'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'เครื่องดื่มที่มีแอลกอฮอล์': 'Alcoholic drinks'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'ยาสูบ หมาก ยานัตถุ์ และอื่นๆ\xa0': 'Tobacco, betel nut, snuff, etc.'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'ที่อยู่อาศัย เครื่องแต่งบ้าน และเครื่องใช้ต่าง ๆ': 'Housing, home decorations and various appliances.'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'เครื่องนุ่งห่มและรองเท้า': 'Clothing and shoes'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'ค่าใช้จ่ายส่วนบุคคล': 'Personal expenses'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'เวชภัณฑ์และค่าตรวจรักษาพยาบาล': 'Medical supplies and medical examination costs'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'การเดินทางและการสื่อสาร': 'Travel and communication'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'การศึกษา': 'Education'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'การบันเทิง การอ่านและกิจกรรมทางศาสนา': 'Entertainment, reading and religious activities'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'การจัดงานพิธีในโอกาสพิเศษ': 'Organizing ceremonies on special occasions'}, inplace=True)
df1['TYPE_EXPENSE'].replace({'ค่าใช้จ่ายที่ไม่เกี่ยวกับการอุปโภคบริโภค': 'Non-consumption expenses'}, inplace=True)
df1
```

```python
def update_plot(type_of_expense):

    df1_filtered = df1[df1['TYPE_EXPENSE'] == type_of_expense]
    
    plt.figure(figsize=(16, 9))
    sns.barplot(data = df1_filtered, x = "TYPE_EXPENSE", y="VALUE" ,hue="YEAR",errorbar=None, palette="rocket_r")
    plt.xlabel('Type of Expense')
    plt.ylabel('Expense')
    plt.title('Expense Sorted by Category')
    plt.legend()
    plt.show()

type_of_expense_values = df1['TYPE_EXPENSE'].unique().tolist()

dropdown_type_of_expense = widgets.Dropdown(options=type_of_expense_values, description='Type of Expense:')
display(dropdown_type_of_expense)

def on_dropdown_change(change):
    update_plot(dropdown_type_of_expense.value)

dropdown_type_of_expense.observe(on_dropdown_change, names='value')

update_plot(type_of_expense_values[0])
```

```python
df2.info()
```

```python
df2['SOCIO_ECONOMIC'].replace({'รวมสถานะทางเศรษฐสังคมของครัวเรือน': 'Household status aggregation'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ถือครองทำการเกษตร การปลูกพืช การเลี้ยงสัตว์ และการเพาะเลี้ยงสัตวน้ำ ซึ่งที่ดินส่วนใหญ่เป็นของตนเอง': 'The landowner engages in agriculture on their own land'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ถือครองทำการเกษตร การปลูกพืช การเลี้ยงสัตว์ และการเพาะเลี้ยงสัตวน้ำ ซึ่งที่ดินส่วนใหญ่เช่าผู้อื่น/ที่สาธารณะ/อื่น ๆ': 'The landowner engages in agriculture on land that leased from others'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ทำการประมง ป่าไม้ ล่าสัตว์ หาของป่า และบริการทางการเกษตร': 'The individual engages in fishing, forestry and agricultural services'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ดำเนินธุรกิจของตนเองที่ไม่ใช่การเกษตร': 'The self-employed individual unrelated to agriculture'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ปฏิบัติงานวิชาชีพ นักวิชาการ และนักบริหารที่รับจ้าง': 'The professional worker is employed on a contractual basis'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'คนงานเกษตร': 'The agricultural laborer'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'คนงานทั่วไป': 'General laborer'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'เสมียนพนักงาน พนักงานขายและให้บริการ': 'Sales and service staff'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ปฏิบัติงานในกระบวนการผลิต': 'Production workers'}, inplace=True)
df2['SOCIO_ECONOMIC'].replace({'ผู้ไม่ได้ปฏิบัติงานในเชิงเศรษฐกิจ': 'Non-economic workforce'}, inplace=True)
df2
```

```python
color = ["lightgray", "lightgray", "red", "red", "red", "red"]

def update_plot(selected_socio_economic, selected_column):
    plt.clf() 

    df_filtered = df2[df2['SOCIO_ECONOMIC'] == selected_socio_economic]

    mean_income_by_socioeconomic = df_filtered.groupby(['SOCIO_ECONOMIC', 'YEAR'])[selected_column].mean().reset_index()

    for socio_economic_group, group_data in mean_income_by_socioeconomic.groupby('SOCIO_ECONOMIC'):
        plt.bar(group_data['YEAR'], group_data[selected_column], label=socio_economic_group, color=color)

    plt.plot(mean_income_by_socioeconomic['YEAR'], mean_income_by_socioeconomic[selected_column], color='blue', marker='o', label='Line Graph')

    plt.xlabel('Year')
    plt.ylabel(f'Mean {selected_column}')
    plt.title(f'Mean {selected_column} by Socio-Economic Group Over the Years')
    plt.show()

socio_economic_values = df2['SOCIO_ECONOMIC'].unique().tolist()

column_values = ['MONTHLY_INCOME', 'CURRENT_INCOME', 'MONEY_INCOME', 'FROM_WORK', 'CURRENT_TRANSFER', 'PROPERTY_INCOME', 'NONMONEY_INCOME', 'NONCUR_M_INCOME']

dropdown_socio_economic = widgets.Dropdown(options=socio_economic_values, description='Socio-Economic:')
dropdown_column = widgets.Dropdown(options=column_values, description='Select Column:')

display(dropdown_socio_economic)
display(dropdown_column)

def on_dropdown_change(change):
    update_plot(dropdown_socio_economic.value, dropdown_column.value)

dropdown_socio_economic.observe(on_dropdown_change, names='value')
dropdown_column.observe(on_dropdown_change, names='value')

update_plot(socio_economic_values[0], column_values[0])
```
