```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = "https://github.com/Schwinnnnn/Data_schwin_year_3/raw/main/Bi/project2/statistic_id794618_real-interest-rates-in-thailand-2013-2022.xlsx"
df = pd.read_excel(url,skiprows=4)
df
```

```python
df = df.drop(df.columns[[0, 3]], axis=1)
df
```

```python
df = df.rename(columns={"Unnamed: 1": "Year", "Unnamed: 2": "Real interest rates(%)"})
df
```

```python
# Assuming df is your DataFrame
sns.set_style("whitegrid")

# Plotting the line plot
sns.lineplot(data=df, x='Year', y='Real interest rates(%)', marker='o', color='blue')

# Adding a horizontal line at y=0
plt.axhline(y=0, color='black', linestyle='-', linewidth=2)

plt.title('')
plt.xlabel('Year')
plt.ylabel('Real interest rates')
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
