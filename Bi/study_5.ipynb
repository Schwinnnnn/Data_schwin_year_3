```python
df.drop(columns=["Brand account"]).copy()
```

```python
@interact
def select_df(i=range(len(dfs))):
    return dfs[i]
#@interact decorator and must not have to below command
select_df(0)
```

```python
@interact
def filter_ig(country=df["Country"].unique()):
    return df[df["Country"] == country]
filter_ig('India')
```

```python
@interact
def filter_usr(regex_pat=""):
    return df[(df["Owner"].str.contains(regex_pat,regex=True,case=False))].sort_values("Followers (millions)",ascending=False)
```
