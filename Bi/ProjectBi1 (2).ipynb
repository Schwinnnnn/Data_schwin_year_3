## <span style="color:lightBLue;"><strong>1. ข้อมูลปริมาณที่ขายในแต่ละประเทศทั่วโลก&nbsp;</strong></span> 


```python
data=pd.read_excel('https://github.com/KNithipon/Projectbi1/raw/main/Data%20real%20marketsize.xls' , skiprows=5)

data
```

```python
year_columns = data.columns[4:]
cumulative_growth = data[year_columns].div(data[year_columns].iloc[:, 0], axis=0) * 100
cumulative_growth['Geography'] = data['Geography'] 
cumulative_growth
melted_data = cumulative_growth.melt(id_vars='Geography', var_name='Year', value_name='Cumulative Growth')
colors = {country: 'red' if country == 'Thailand' else 'lightgrey' for country in melted_data['Geography'].unique()}
melted_data
# sns_plot = sns.lineplot(data=melted_data, x='Year', y='Cumulative Growth', hue='Geography', palette=colors)

```

## <span style="color:lightBLue;"><strong>2. ข้อมูลYear on Year(%)ที่ขายในแต่ละประเทศทั่วโลก&nbsp;</strong></span> 


```python
df1=pd.read_excel('https://github.com/KNithipon/Projectbi1/raw/main/Yoy.xlsx')
df1.head()
```

```python
plt.figure(figsize=(10, 6)) 
colors = {country : 'red' if country == 'Thailand' else 'lightgray' for country in df1['Geography'].unique()}

sns.lineplot(data=df1, x='Year', y='YoY', hue='Geography', palette=colors, )
plt.title('Soft drink low sugar')
plt.show()
```

## <span style="color:lightBLue;"><strong>3. Brand Sharesของแต่ละบริษัทในประเทศไทย ในปี2017&nbsp;</strong></span> 


```python
df2=pd.read_excel('https://github.com/KNithipon/Projectbi1/raw/main/Passport_Stats_28-01-2024_1626_GMT.xls',skiprows=5)

df2
```

```python
data_for_pie = df2.groupby('Brand Name')['2017'].sum()

plt.figure(figsize=(8, 8))
plt.title("Brand market size in 2017")
color = ["red","lightgray","lightblue","gray","purple"]

plt.pie(data_for_pie, labels=data_for_pie.index, autopct='%1.1f%%', startangle=140,colors=color )
plt.axis('equal')  
plt.show()
```
