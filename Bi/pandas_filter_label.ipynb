## ตัวอย่างที่ 1: ชื่อคอลัมน์ที่เป็นตัวเลข

```python
url='https://github.com/prasertcbs/basic-dataset/raw/master/iphone%20sales%20in%20asia.xlsx'
df=pd.read_excel(url, index_col='Geography')
df
```

```python
df.columns
```

```python
df[2015]
```

```python
df[2015:]
```

```python
df.columns >= 2015
```

```python
df[df.columns[df.columns >= 2015]] # boolean index
```

```python
df.filter(regex='[6-9]$', axis=1)
```

```python
df.filter(regex='(0|3|5)$', axis=1)
```

```python
df.filter(items=[2015, 2019], axis=1)
```

```python
df.filter(regex='Thailand', axis=0)
```

```python
df.filter(regex='Thailand|Vietnam', axis=0)
```

```python
df.filter(regex='^(T|V|C)', axis=0)
```

```python
df.filter(regex='(d|a)$', axis=0)
```

```python
df.filter(regex='sia$', axis=0)
```

```python
df.filter(regex='sia$', axis=0).filter(regex='[5-9]$', axis=1)
```

## ตัวอย่าง 2

```python
ds=pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/score_a.tsv',
                index_col='studentid')
ds
```

```python
ds.filter(regex='^math', axis=1)
```

```python
ds.filter(regex='2$', axis=1)
```

```python
ds.filter(regex='^(math|eng)', axis=1)
```

```python
ds.filter(regex='^(math|eng)2', axis=1)
```

```python
ds.columns.str.match('^(math|eng)2')
```

```python
ds[ds.columns[ds.columns.str.match('^(math|eng)2')]]
```

```python
ds.filter(regex='^(math|eng)2', axis=1).filter(regex='^A', axis=0)
```
